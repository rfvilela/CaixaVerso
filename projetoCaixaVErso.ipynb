```python
from datetime import datetime
import json


catalogo = []

while True:
    while True:
        nomeFilme = input("Digite o Nome do(a) Filme/Série: ")
        if nomeFilme:
            break
        print("O nome do(a) Filme/Série não pode estar vazio")

    while True:
        nomeDiretor = input("Digite o Nome do Diretor do(a) Filme/Série: ")
        if nomeDiretor:
            break
        print("O nome do Diretor do(a) não pode estar vazio")

    while True:
        avaliacao = float(input("Avaliação (0 a 10): "))
        if 0 <= avaliacao <= 10:
            break
        else:
            print("Digite um valor entre 0 e 10.")

    while True:
        data_str = input("Data de lançamento (dd/mm/aaaa): ")
        try:
            data_lancamento = datetime.strptime(data_str, "%d/%m/%Y").date()
            break
        except ValueError:
            print("Formato inválido. Use dd/mm/aaaa.")

    while True:
        try:
            horas = int(input("Duração - horas: "))
            minutos = int(input("Duração - minutos: "))
            if horas >= 0 and 0 <= minutos < 60:
                duracao = (horas, minutos)
                break
            else:
                print("Digite horas ≥ 0 e minutos entre 0 e 59.")
        except ValueError:
            print("Digite valores numéricos.")
    while True:
        disponivel_str = input("Disponível no streaming? (s/n): ").strip().lower()
        if disponivel_str in ["s", "n"]:
            disponivel = disponivel_str == "s"
            break
        print("Responda apenas com 's' ou 'n'.")

    filme = {
        "nome": nomeFilme,
        "diretor": nomeDiretor,
        "avaliacao": avaliacao,
        "data_lancamento": data_lancamento.strftime("%Y-%m-%d"),
        "duracao": duracao,
        "disponivel_streaming": disponivel
    }
    catalogo.append(filme)

    continuar = input("Deseja cadastrar mais um filme? (s/n): ").strip().lower()
    if continuar == "n":
        break



#exemplo na aula do dia 14/08/2025###
with open("catalogo.json", "w") as arq:
    json.dump(catalogo, arq)


#teste github
```

```python
from functools import reduce

def extrair_avaliacao(filme):
    return filme["avaliacao"]

def somar(x, y):
    return x + y


#exemplo na aula do dia 14/08/2025###
with open("catalogo.json", "r") as arq:
    string_json = arq.read()
    catalogo = json.loads(string_json)


#dica na aula do dia 21/08 usando a função map
avaliacoes = list(map(extrair_avaliacao, catalogo))

#dica na aula do dia 21/08 usando a função reduce
soma = reduce(somar, avaliacoes)
media = soma / len(avaliacoes)


avaliacoes_ordenadas = sorted(avaliacoes)
n = len(avaliacoes_ordenadas)

if n % 2 == 1:
    mediana = avaliacoes_ordenadas[n // 2]
else:
    mediana = (avaliacoes_ordenadas[n // 2 - 1] + avaliacoes_ordenadas[n // 2]) / 2


print(f"Média das avaliações: {media:.2f}")
print(f"Mediana das avaliações: {mediana:.2f}")


def filmes_com_avaliacao_maior_que(lista_filmes=catalogo, nota_minima=7):
    filmes_filtrados = [
        filme["nome"]
        for filme in lista_filmes
        if filme["avaliacao"] > nota_minima
    ]
    return filmes_filtrados


filmes_bons = filmes_com_avaliacao_maior_que()
print("Filmes com avaliação > 7:", filmes_bons)


filmes_top = filmes_com_avaliacao_maior_que(nota_minima=8)
print("Filmes com avaliação > 8:", filmes_top)



filmes_streaming = [
    filme["nome"]
    for filme in catalogo
    if filme["disponivel_streaming"] is True
]

print("Filmes disponíveis em streaming:", filmes_streaming)



def filmesMaiores(lista_filmes=catalogo):
    if not lista_filmes:
        return None, None

    #Dica do professor - transofrmar tudo em minutos#
    def duracaoMin(duracao):
        horas, minutos = duracao
        return horas * 60 + minutos

    
    filme_maior = lista_filmes[0]
    filme_menor = lista_filmes[0]

    for filme in lista_filmes[1:]:
        if duracaoMin(filme["duracao"]) > duracaoMin(filme_maior["duracao"]):
            filme_maior = filme
        if duracaoMin(filme["duracao"]) < duracaoMin(filme_menor["duracao"]):
            filme_menor = filme

    return filme_maior, filme_menor


maior, menor = filmesMaiores()
print("Filme mais longo:", maior["nome"], "com duração: ", maior["duracao"])
print("Filme mais curto:", menor["nome"], "com duração: ", menor["duracao"])


contagem_diretores = {}




for filme in catalogo:
    diretor = filme["diretor"]
    if diretor in contagem_diretores:
        contagem_diretores[diretor] += 1
    else:
        contagem_diretores[diretor] = 1




max_filmes = 0
moda_diretores = []



for diretor, qtd in contagem_diretores.items():
    if qtd > max_filmes:
        max_filmes = qtd
        moda_diretores = [diretor]
    elif qtd == max_filmes:
        moda_diretores.append(diretor)




if moda_diretores:
    print(f"Diretores que mais apareceram: {max_filmes} {', '.join(moda_diretores)}")
else:
    print("Nenhum diretor encontrado.")
```

```python
import csv

listaNotas = [[6.3, 2.1, [8.4, 4.2, 5.1], 9.6], 10, 4.7, 6.5]

def soma(lista):
    total = 0
    for item in lista:
        if isinstance(item, list):
            total += soma(item)
        else:
            total += item
    return total


def contagem(lista):
    cont = 0
    for item in lista:
        if isinstance(item, list):
            cont += contagem(item)
        else:
            cont += 1
    return cont


def media(lista):
    return soma(lista) / contagem(lista)


media_turma = media(listaNotas)
print("A Média da turma é:", round(media_turma, 2))

#lista de notas desejadas - poderia solicitar via input, mas preferi escolher uma aleatória
notas_desejadas = [6.0, 5.5, 7.0, 4.5, 8.0]


def ajustar_notas(notas_desejadas, media_desejada):
    soma = 0
    for nota in notas_desejadas:
        soma += nota
    media_atual = soma / len(notas_desejadas)

    fator = media_desejada / media_atual

    novas_notas = []
    for nota in notas_desejadas:
        novas_notas.append(nota * fator)
    return novas_notas


novas_notas = ajustar_notas(notas_desejadas, media_turma)
print("Novas notas ajustadas:", end=" ")
notas_arredondadas = []
for n in novas_notas:
    notas_arredondadas.append(round(n, 2))
print(notas_arredondadas)


soma = 0
for n in novas_notas:
    soma += n
nova_media = soma / len(novas_notas)
print("Nova média:", round(nova_media, 2))


with open("notas_ajustadas.csv", "w") as arquivo_csv:
    escritor = csv.writer(arquivo_csv)
    escritor.writerow(["Aluno", "Nota"])
    for i in range(len(novas_notas)):
        escritor.writerow([f"Aluno {i+1}", round(novas_notas[i],2)])
```
